Intención - Hacer calzar un objeto que no controlamos a nuestra interfaz 

Problema - Sistema tiene datos y comportamiento correctos pero interfaz incorrecta 

Solución - Adaptador provee un envoltorio con interfaz deseada 

Implementación - incorporar la clase existente en otra que tiene la interfaz deseada y hacer que los métodos de adaptador llamen a métodos del adaptado

```ruby
class Hammer
    def swing!
      puts ("swinging the hammer")
    end
  end
  
  class Tool
    def initialize(adapter)
      @adapter = adapter
    end
    
    def use_tool
      @adapter.use_tool
    end
  end
```

Definimos la clase que adapta estas dos herramientas, inicializa una y crea un método para que dicha clase pueda usarse con el método de otra clase

```ruby
class HammerAdapter
    def initialize(hammer)
      @hammer = hammer
    end
    def use_tool
      @hammer.swing!
    end
  end
```

```ruby
  hammer = Hammer.new
  adapter = HammerAdapter.new(hammer)
  
  tool = Tool.new(adapter)
  tool.use_tool
  
```
